#Passo 2: Criar o script da aplicação (app.py)

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
from PIL import Image

# 1. Configuração da Página
st.set_page_config(
    page_title="Northwind Persistência Poliglota",
    page_icon="/content/NortPersPoliglota.png",
    layout="wide"
)

# Estilo CSS para melhorar a estética (Opcional)
st.markdown("""
    <style>
    .main {
        background-color: #f5f7f9;
    }
    </style>
    """, unsafe_allow_html=True)

# 2. Cabeçalho e Identidade Visual
st.title("Northwind: Persistência Poliglota")
st.markdown("### Implementação e Análise Comparativa de Persistência Poliglota sobre o Domínio Northwind Traders.")
st.write("---")

# Espaço para o Logo (Se você tiver o arquivo no Colab)
try:
    logo = Image.open("/content/NortPersPoliglota.png") # Correção aqui
    st.sidebar.image(logo, width='stretch') # Correção aqui
except:
    st.sidebar.header("Northwind: Persistência Poliglota")

# 3. Sidebar - Navegação
with st.sidebar:
    st.header("Menu de Navegação")
    st.image("/content/NortPersPoliglota.png", width='stretch') # Correção aqui

    selected = st.radio(
        "Selecione o Módulo:",
        ["Dashboard Geral", "Análise de Persistência", "Consultas SQL", "Documentos NoSQL"]
    )
st.sidebar.title("Navegação")
menu = st.sidebar.radio("Selecione uma seção:",
    ["Visão Geral", "Análise Comparativa", "Persistência SQL", "Persistência NoSQL (JSON)"])

# 4. Conteúdo das Seções
if menu == "Visão Geral":
    st.header("Domínio Northwind Traders")
    st.write("""
    Este Projeto Integrador Solução de Inteligência Artificial Baseada em Dados, focado na Área 3 - Banco de Dados, propõe uma análise de persistência poliglota sobre o domínio Northwind Traders. O objetivo geral é implementar e comparar rigorosamente soluções relacionais (PostgreSQL) e baseadas em documentos (MongoDB), avaliar critérios de normalização, escalabilidade e performance. O cronograma do projeto está estruturado em quatro entregas críticas: o kick-off com modelagem conceitual; a implementação técnica de scripts e migrações; o benchmark de consultas complexas através de mais de 15 (quinze) Aggregation Pipelines e queries SQL; e, por fim, a consolidação de um guia de decisão arquitetural.
Para viabilizar a execução deste projeto em um ciclo de sete semanas foi adotada a metodologia ágil Scrum como framework de gestão. O Scrum é essencial neste contexto para organizar o backlog de tarefas técnicas e garantir entregas incrementais através de Sprints semanais. A utilização de métodos ágeis permite realizar alinhamentos rápidos, identificar impedimentos na migração de dados e ajustar as prioridades conforme os resultados dos testes de performance, para garantir que o relatório final seja entregue sem atrasos.
O desenvolvimento estratégico é guiado pela metodologia CRISP-DM (Cross Industry Standard Process for Data Mining) para atuar como um processo padrão que permite prever problemas e criar soluções a partir de informações já existentes. Essa metodologia será executada através de suas seis etapas fundamentais:
- Entendimento do Negócio: Definir os objetivos do projeto e as necessidades da Northwind.
- Entendimento dos Dados: Iniciar a exploração das fontes de dados, questionar sua estruturação, formatos e acessibilidade, para garantir uma coleta integral e sem perdas de informações críticas.
- Preparação dos Dados: Tratar valores nulos, fusão de dados e garantia de formatos corretos para minimizar drasticamente o retrabalho nas fases posteriores.
- Modelagem: Implementar o esquema relacional e o design de documentos NoSQL, observando como os dados se comportam em cada estrutura.
- Avaliação: Avaliar os resultados de performance e a facilidade de consulta correspondem às expectativas. Caso negativo, as forças são direcionadas para correções nos atributos ou na entrada de dados.
- Implementação (Deployment): Colocar em produção, agregar valor real ao negócio através de recomendações arquiteturais e inteligência de mercado.
É importante ressaltar que o CRISP-DM é uma metodologia extremamente versátil, pois integra a criação de modelos técnicos ao entendimento profundo do negócio. No entanto, é necessário compreender que este processo não é estático; ao final das seis etapas, o ciclo provavelmente precisará ser reiniciado e ajustado conforme novos dados surjam, permitindo que a solução de inteligência artificial evolua continuamente. Portanto, é através desta abordagem híbrida e cíclica, que o projeto busca definir a ferramenta ideal para cada carga de trabalho, para culminar em um relatório técnico completo e estratégico.

    """)


elif menu == "Análise Comparativa":
    st.header("Comparativo de Performance")
    st.write("Resultados da análise entre SQL vs NoSQL.")

    # Exemplo de gráfico comparativo
    chart_data = pd.DataFrame(
       np.random.randn(20, 2),
       columns=['SQL (PostgreSQL)', 'NoSQL (MongoDB)'])
    st.line_chart(chart_data)

elif menu == "Persistência SQL":
    st.header("Persistência Relacional")
    st.info("Focado em integridade referencial e consultas complexas (JOINs).")

elif menu == "Persistência NoSQL (JSON)":
    st.header("Persistência Orientada a Documentos")
    st.success("Focado em escalabilidade e esquemas flexíveis.")

# Conteúdo Principal
if selected == "Dashboard Geral":
    col1, col2, col3 = st.columns(3)
    col1.metric("Bases de Dados", "SQL + NoSQL")
    col2.metric("Domínio", "Northwind Traders")
    col3.metric("Status", "Em Desenvolvimento")

    st.write("### Sobre o Projeto")
    st.info("Este projeto analisa como o domínio Northwind se comporta em um ambiente poliglota.")

elif selected == "Análise de Persistência":
    st.write("### Comparativo de Performance: Escrita vs Leitura")
    # Gerando dados fictícios para demonstração
    chart_data = pd.DataFrame(
        np.random.rand(10, 2),
        columns=['Latência SQL (ms)', 'Latência NoSQL (ms)']
    )
    st.area_chart(chart_data)

elif selected == "Consultas SQL":
    st.write("### Exploração Relacional (SQLite/PostgreSQL)")
    st.code("SELECT * FROM Orders JOIN Customers ON Orders.CustomerID = Customers.CustomerID LIMIT 5;", language='sql')
    st.table(pd.DataFrame({'OrderID': [10248, 10249], 'Customer': ['Vins et alcools', 'Toms Spezialitäten']}))

elif selected == "Documentos NoSQL":
    st.write("### Persistência de Documentos (JSON/MongoDB)")
    st.json({
        "order_id": 10248,
        "customer": "Vins et alcools",
        "items": [
            {"product": "Queso Cabrales", "qty": 12},
            {"product": "Singaporean Hokkien Fried Mee", "qty": 10}
        ]
    })
```
